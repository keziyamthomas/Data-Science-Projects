```python
# convert postal code column to INteger type as was in original dataset
superstore_df['Postal Code'] = superstore_df['Postal Code'].astype('int64',copy=False)
```

```python
# filtering all rows corresponding to Burlington City and state of Vermont to check if any of them have Postal
# code column filled
superstore_df.loc[(superstore_df['City'] == "Burlington") & (superstore_df['State'] == "Vermont")]
```

```python
# There are two methods to fill postal code column when it's empty, the former is more efficient when all rows
# together in one go. If one of these rows are already filled with the correct value, you can use the transform
# function to fill the rest of the rows with the same value

superstore_df.loc[(superstore_df['City'] == "Burlington") & (superstore_df['State'] == "Vermont"), ['Postal Code']] = "05401"

# 0R 

superstore_df.at[9146,'Postal Code'] = "05401"
superstore_df['Postal Code'] = superstore_df['Postal Code'].fillna(superstore_df.groupby(['City','State'])['Postal Code'] \
                                                                   .transform('first'))
```

```python
# checking the rows to see if Postal code is updated
superstore_df.loc[(superstore_df['City'] == "Burlington") & (superstore_df['State'] == "Vermont")]
```
